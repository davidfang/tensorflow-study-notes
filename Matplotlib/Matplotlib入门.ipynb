创建一个图像的最简单方法是使用pylot：

```python
fig = plt.figure()
fig.suptitle("No axes on this figure")

fig,ax_lst = plt.subplots(2,2)
```

```python
x = np.linspace(0,2,100)

plt.plot(x,x,label='linar')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')

plt.xlabel('X label')
plt.ylabel('Y label')

plt.title('Simple plot')

plt.legend()

plt.show()
```

```python
x = np.arange(0,10,0.2)
y = np.sin(x)
fig,ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
data = {
    'a':np.arange(50),
    'c':np.random.randint(0,50,50),
    'd':np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a','b',c='c',s='d',data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0,5.0,0.01)
s = np.cos(2*np.pi*t)
line,=plt.plot(t,s,lw=2)

plt.annotate('local max',xy=(2,1),xytext=(3,1.5),
            arrowprops=dict(facecolor='black',shrink=0.05),)

plt.ylim(-2,2)
plt.show()
```

```python
# 普林斯顿微积分  66页 连续函数
fig = plt.figure(figsize=(24,8))
ax = plt.axes()

x = np.linspace(-0.3,0.3,1000)
# ax.plot(x,np.sin(x))


# x = np.arange(-1,1,0.001)
y = np.sin(1/x) * x
# fig,ax = plt.subplots()

ax.axvline(0)
ax.axhline(0)
ax.plot(x,y)
ax.plot(x,x,'--')
ax.plot(x,-x,'--')
plt.show()
```

```python
# 微积分  连续函数和倒数
fig = plt.figure(figsize=(24,8))
ax = plt.axes()

x = np.linspace(-3,3,1000)
# ax.plot(x,np.sin(x))


# x = np.arange(-1,1,0.001)
y = np.square(x) #平方
z = x * 2 #倒数
# fig,ax = plt.subplots()


ax.axvline(0)
ax.axhline(0)
ax.plot(x,y)
ax.plot(x,z,'--')
# ax.plot(x,-x,'--')
plt.show()
```

### 普林斯顿微积分  112页 三角函数的倒数

$$ \lim_{x \to 0} \frac{sin(x)}{x} = 1 $$

```python
# 普林斯顿微积分  112页 三角函数的倒数
fig = plt.figure(figsize=(24,8))
ax = plt.axes()

x = np.linspace(-1,1,100)
# ax.plot(x,np.sin(x))


# x = np.arange(-1,1,0.001)
y = np.sin(x) 
# fig,ax = plt.subplots()

ax.axvline(0)
ax.axhline(0)
ax.plot(x,y)
ax.plot(x,x,'--')
# ax.plot(x,-x,'--')
plt.show()
```

```python
# 普林斯顿微积分  124页 三角函数极限和导数
fig = plt.figure(figsize=(24,8))
ax = plt.axes()

x = np.linspace(-100,100,1000)
# ax.plot(x,np.sin(x))


# x = np.arange(-1,1,0.001)
y = np.sin(x) / x
# fig,ax = plt.subplots()

ax.axvline(0)
ax.axhline(0)
ax.plot(x,y)
# ax.plot(x,1/x,'--')
# ax.plot(x,-1/x,'--')
plt.show()
```
