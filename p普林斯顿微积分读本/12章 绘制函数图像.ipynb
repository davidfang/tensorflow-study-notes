## 第12章 绘制函数图像

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from math import pi,cos,sin,tan
```

```python
# 定义变量
x = np.linspace(-5,5,200)
x = x[x!=0]
x = x[x!=-2]
y = (x-3)*(x-1)**2 / (x**3 * (x+2))

# 过滤掉不符合定义域的 `x` 值
# 过滤掉 `x = 0` 和 `x = -2` 的点
# y[x == 0] = np.nan
# y[x == -2] = np.nan

# 绘制图像


# 绘制图像
plt.figure(figsize=(24,8))
# plt.style.use('seaborn-whitegrid') # 设置使用的样式
# plt.style.use('ggplot')
plt.style.use('seaborn')
# plt.style.use('bmh')




plt.xlabel('x')
plt.ylabel('y')

# 设置 `x` 和 `y` 轴的范围
# plt.xlim([0, 3])
plt.ylim([-20, 20])


# plt.fill([0, -2], [np.inf, np.inf], 'gray')
# 添加网格
# plt.grid(visible=False)

# 设置图像的标题
plt.title(r'$y = \frac{(x-3)(x-1)^2}{x^3(x+2)}$')

# 添加图例
# plt.legend([r'$y = \frac{(x-3)(x-1)^2}{x^3(x+2)}$'])


# plt.plot(x[x != 0], y[x != 0])
plt.plot(x,y)
# plt.scatter(x,y)
# 设置点的大小和颜色
# plt.scatter(x[x != 0], y[x != 0], s=10, c='red')

# 添加点
# plt.scatter([0.5, 1], [0.5, 2], color='red', s=5)

# 保存图像
# plt.savefig('my_plot.png')

plt.show()
```
