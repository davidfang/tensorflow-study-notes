### 第2章 二维向量绘图

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from math import pi,cos,sin,tan
```

#### 2.3.4 练习

```python
#一个从笛卡儿坐标到极坐标的转换器。 
def to_polar(vector): 
    x, y = vector[0], vector[1] 
    angle = math.atan2(y,x) 
    return (math.length(vector), angle)
#接收一对极坐标并返回相应的笛卡儿坐标。
def to_cartesian(polar_vector): 
    length, angle = polar_vector[0], polar_vector[1] 
    return (length*cos(angle), length*sin(angle))

```

```python
# 练习2.37

polar_coords = [(cos(5 * x * pi/500), 2 * pi * x/1000) for x in range(0, 1000)]
vectors = [to_cartesian(p) for p in polar_coords]
fig = plt.figure(figsize=(8,8))
ax = plt.axes()

x = [x[0] for x in vectors]
y = [y[1] for y in vectors]
ax.scatter(x,y)

plt.show()
```
